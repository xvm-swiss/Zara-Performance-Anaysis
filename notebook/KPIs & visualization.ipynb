## KPIs Section

#### 1) Total No. of Transactions
#### 2) No. of terms
#### 3) No. of Units Sold
#### 4) Avg. Price
#### 5) Total Sales 

```python
no_of_transactions = len( df )
no_of_terms = df['terms'].nunique()
no_of_units_sold = int( df['Sales_Volume'].sum() )
avg_price = float( round( df['price'].mean(), 2 ) )
total_sales = float( round( df['Total_Sales'].sum(), 2 ) )
```
